```python
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
```

```python
b_cancer = load_breast_cancer()
```

```python
data = pd.DataFrame(b_cancer.data, columns=b_cancer.feature_names)
target = pd.DataFrame(b_cancer.target, columns=['target'])
```

```python
import seaborn as sns
```

```python
b_df = pd.concat([data, target], axis=1)
```

```python
b_df
```

```python
def remove_outliers(df, column_name):
    s = df[column_name]
    sd = s.std()
    mean = s.mean()
    min_val = mean - 3 * sd
    max_val = mean + 3 * sd
    
    return df[(s < max_val) & (s > min_val)]

b_df = remove_outliers(b_df, 'mean area')
```

```python
b_df = remove_outliers(b_df, 'w')
```

```python
outlierb_df = b_df.copy()
```

```python
outlierb_df.reset_index(drop=True, inplace=True)
```

```python
outlierb_df
```

```python
import matplotlib.pyplot as plt

# 모든 숫자형 칼럼에 대해 boxplot 그리기
plt.figure(figsize=(10, 6))  # 그림 크기 설정

b_df.boxplot(column=list(b_df.select_dtypes(include=['number']).columns))
plt.xticks(rotation=45)  # x축 눈금 레이블 회전하여 표시


plt.tight_layout()  # 레이아웃 조정
plt.show()
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
features = outlierb_df.drop('target', axis=1)
scaled_features = scaler.fit_transform(features)

scaled_df = pd.DataFrame(scaled_features, columns=features.columns)
scaled_df = pd.concat([scaled_df, outlierb_df['target']], axis=1)
```

```python
X = scaled_df
Y = outlierb_df["target"]
```

```python
from sklearn.model_selection import cross_val_predict
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report
from sklearn.linear_model import LogisticRegression

# 로지스틱 회귀 모델 생성
logistic_model = LogisticRegression()

# k-fold 교차검증을 통한 예측
y_pred = cross_val_predict(logistic_model, X, Y, cv=10)  # 5-fold 교차검증 예시

# 오차 행렬
conf_matrix = confusion_matrix(Y, y_pred)
print("오차 행렬:\n", conf_matrix)

# 정확도
accuracy = accuracy_score(Y, y_pred)
print("정확도:", accuracy)

# 정밀도
precision = precision_score(Y, y_pred)
print("정밀도:", precision)

# 재현율
recall = recall_score(Y, y_pred)
print("재현율:", recall)

# F1-score
f1 = f1_score(Y, y_pred)
print("F1-score:", f1)

# AUC
y_pred_proba = cross_val_predict(logistic_model, X, Y, cv=5, method='predict_proba')
auc = roc_auc_score(Y, y_pred_proba[:, 1])
print("AUC:", auc)

print(classification_report(Y,y_pred,labels =  [1,0]))
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import confusion_matrix

# 실제값과 예측값으로 오차 행렬 생성
conf_matrix = confusion_matrix(Y, y_pred)  # y_true: 실제값, y_pred: 예측값
sns.set(font_scale=1.4)  # 전체 폰트 크기 설정
# 오차 행렬 히트맵 시각화
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 로지스틱 회귀 모델 생성
logistic_model = LogisticRegression()

# k-fold 교차검증을 통한 예측
y_pred_proba = cross_val_predict(logistic_model, X, Y, cv=10, method='predict_proba')

# AUC 계산
auc = roc_auc_score(Y, y_pred_proba[:, 1])
print("AUC:", auc)

# ROC curve 계산
fpr, tpr, thresholds = roc_curve(Y, y_pred_proba[:, 1])

# ROC curve 그리기
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f'AUC = {auc:.2f}')
plt.plot([0, 1], [0, 1], 'r--', label='Random Guessing')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve')
plt.legend()
plt.grid(True)
plt.show()
```

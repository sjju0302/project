```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
titanic = pd.read_csv("clean_titanic.csv" ,index_col = 0)
```

```python
titanic_male = titanic[titanic["sex"] == "male"]
titanic_female = titanic[titanic["sex"] == "female"]
```

```python
x = [titanic_male[titanic_male["survived"] == 1]["survived"].count(),titanic_male[titanic_male["survived"] == 0]["survived"].count()]
y = ["survived","death"]
z = [titanic_female[titanic_female["survived"] == 1]["survived"].count(),titanic_female[titanic_female["survived"] == 0]["survived"].count()]
a = [titanic_male[titanic_male["survived"] == 1]["survived"].count(),titanic_female[titanic_female["survived"] == 1]["survived"].count()]
```

```python

colors = ['#ADD8E6', '#696969']
plt.subplot(1, 3, 1)
plt.pie(x, labels=y, colors=colors,autopct='%1.1f%%')
plt.title("Male")

plt.subplot(1, 3, 2)
plt.pie(z, labels=y, colors=colors,autopct='%1.1f%%')
plt.title("Female")

plt.subplot(1, 3, 3)
plt.pie(a, labels = ["male","female"], colors=["tab:blue","tab:pink"],autopct='%1.1f%%')
plt.title("Difference")
plt.tight_layout()  
plt.show()
```
